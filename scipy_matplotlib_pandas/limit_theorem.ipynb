```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from statsmodels.distributions.empirical_distribution import ECDF
%matplotlib inline
```

#                          Центральная предельная теорема

### Гамма распределение: $$f(x, a) = \lambda^{a}x^{(a-1)}\frac{e^{-\lambda x}}{\gamma(a)}$$
#### Почему я её выбрал? Ну в ней присуствует гамма-функция Эйлера. Обобщенные функции по матану будем проходить в конце семестра и я решил ознакомится как выглядит распределение на основе данной функции.
#### Математическое ожидание случайной величины x : $$E[x] = \frac{a}{\lambda} $$
#### Дисперсия случайной величины x: $$D[x] = \frac{a}{\lambda^2} $$
#### В нашем случае $$ \lambda = 1 $$

```python
a = 1.99 # параметр гамма распределения
D = a
E = a
rv = sts.gamma(a) #  гамма распределение
r = sts.gamma.rvs(a, size=1000) # создаем выборку гамма распределния размером 1000, параметр а
```

### Получается что $$ E[x] = D[x] = 1.99 $$

```python
x = np.linspace(0, 6, 1000)

#Строим плотность распределения и гистограмму
plt.plot(x, rv.pdf(x), label='theor gamma')

plt.hist(r, density=True, label = 'gamma hist')
plt.legend(loc='best', frameon=False)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

# Оценим распределение выборочного среднего случайной величины при разных объёмах выборок

# n = 5

```python
n = 5

D5 = D/n #дисперсия для нормального распределения приближающего выборку средних
sigma5 = np.sqrt(D5) #сигма для нормального распределения выборки средних
norm5 = sts.norm(E, sigma5) #нормальное распределение

sample5=np.array([])#обозначим массив выборочных средних
i = 1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    sample5i = sts.gamma.rvs(a,size = n)#генерация выборки
    mean5i = np.sum(sample5i)/n#определение выборочного среднего
    sample5 = np.append(sample5,mean5i)#добавляем в массив новое значение
    i+=1

x = np.linspace(0, 5, 1000)
pdf5 = norm5.pdf(x)

plt.plot(x, pdf5, label='PDF(n=5)')
plt.hist(sample5, bins = 20, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
```
